```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d
#parameters, coordinates and fields
pi=np.pi
labels = 'Winter', 'Spring', 'Summer', 'Autumn'
sizes = [14, 20, 25, 31]
explode = (0, 0.1, 0, 0) # only "explode" the 2nd slice (i.e. ’Spring’)
explode = (0, 0.1, 0, 0) # only "explode" the 2nd slice (i.e. ’Spring’)
nlon =361
lon=np.linspace(0,360,nlon)
nlat=181
lat=np.linspace(-90,90,nlat)
X,Y=np.meshgrid(lon,lat)
T=270*(np.cos(Y*pi/180)**2)*(np.cos(X*pi/180)**2)+100.0
U=-10*(np.sin(2*Y*pi/180))*(np.cos(X*pi/180)**2)
V=10*(np.cos(Y*pi/180)**2)*(np.sin(2*X*pi/180))
fig = plt.figure(figsize=(14,10))
ax1=fig.add_subplot(3,2,1)
ax1.pie(sizes, explode=explode, labels=labels)
ax1.set_title('Subplot 1: Pie chart')
ax2=fig.add_subplot(3,2,2)
ax2.bar(labels,sizes)
ax2.set_title('Subplot 2: Bar chart')
ax3=fig.add_subplot(3,2,3)
colmesh=ax3.pcolormesh(X,Y,T, cmap="coolwarm", vmin=100, vmax=350)
plt.colorbar(colmesh,ax=ax3)
ax3.set_title('Plot 3: colormesh')
ax4=fig.add_subplot(3,2,4)
step=10
qv=ax4.quiver(X[::step,::step],Y[::step,::step],U[::step,::step],
V[::step,::step],scale=200.0, color='red')
q_typ=10.0
Qkey1 = plt.quiverkey(qv, 0.95,-0.12 , q_typ, '%.1f m/s' % q_typ, coordinates='axes')
ax4.set_title('Plot 4: quiver')
ax5=fig.add_subplot(3,2,5)
# find the row corresponding to the equator in data
ind=np.where(abs(Y[:,0])<0.1) #
indeq=ind[0] #
#find the row corresponding to 30 N
ind=np.where(abs(Y[:,0]-30)<0.1) #
ind30n=ind[0]
#find the row corresponding to 45N
#find the row corresponding to 45N
ind=np.where(abs(Y[:,0]-45)<0.1) #
ind45n=ind[0]
ax5.plot(X[indeq,:].T, T[indeq,:].T, label='Equator')
ax5.plot(X[ind30n,:].T, T[ind30n,:].T, label='30N')
ax5.plot(X[ind45n,:].T, T[ind45n,:].T, label='45N')
ax5.legend()
ax5.set_title('Plot 5: Temperature at Equator, 30N and 45N' )
ax6=fig.add_subplot(3,2,6,projection='3d')
ax6.plot_surface(X,Y,T)
ax6.set_title('Plot 6: Temperature surface plot')
fig.suptitle('Subplot demo')
```

```python
from numba import jit
import numpy as np
x = np.arange(100).reshape(10, 10)
@jit(nopython=True) # Set "nopython" mode for best performance, equivalent to @njit'

def go_fast(a): # Function is compiled to machine code when called the first time
    trace = 0.0
    for i in range(a.shape[0]): # Numba likes loops
    trace += np.tanh(a[i, i]) # Numba likes NumPy functions
    return a + trace
# Numba likes NumPy broadcasting
print(go_fast(x))
```
