```python
import numpy as np
import scipy
import seaborn as sns

np.random.seed(3)
X=np.random.randint(1,7,size=10000)
# generate 10000 random integers between 1 and 7 (excluded)
sns.histplot(x=X)
```

```python
sns.histplot(x=X, discrete=True, stat='probability')
```

```python
sns.histplot(x=X, discrete=True, stat='density', kde=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

X = np.random.normal(size=100000)
ax1 = plt.subplot(2,1,1)
sns.histplot( X, ax=ax1, stat="probability")
ax2 = plt.subplot(2,1,2)
sns.histplot( X, ax=ax2, stat="density", kde=True)
sns.kdeplot(X, ax=ax2, color="red")
```

```python
sns.boxplot(x=X)
sns.violinplot(x=X)
```

```python
scipy.stats.moment(X, moment=3)
```

```python
scipy.stats.skew(X)
```

```python
import xarray as xr
import pandas as pd
import seaborn as sns
fname= '2000monthly-meant.nc'
ds = xr.open_dataset(fname, engine='scipy')
tsea=ds.t.sel(level=1000.0, method="nearest")
df=tsea.to_dataframe()
# convert latitude and time to columns
df1=df.reset_index(level=['latitude', 'time'])
sns.scatterplot(data=df1,x="latitude",y="t",hue="time")
```

```python
sns.scatterplot(x=df1["latitude"],y=df1["t"],hue=df1["time"])
```
